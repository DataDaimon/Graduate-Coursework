```python
import pandas
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
```

```python
df = pandas.read_csv('/Users/Richard/OneDrive - Western Governors University/MSDA/Databases/Churn DB/churn_raw_data.csv')
```

### Verify variable types

```python
data = df.dtypes
print(data)
```

### Displaying unique values

```python
data = df['Interaction'].value_counts()
print(data)
```

### Detecting Duplicate Values

```python
data = df.loc[df.duplicated()]
print(data)
```

### Detecting Missing Data

```python
data = df.isnull().sum()
print(data)
```

### Identifying Standard Deviation

```python
data = df.std()
print(data)
```

### Defining Outliers

```python
outliers_z = (df['Population'] - df['Population'].mean()) /df['Population'].std()
data = df.loc[(outliers_z > 3) | (outliers_z < -3)]
print(data)
```

### Missing Data

```python
cols = df.columns[:30] # first 30 columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colours))

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

### Z-Scores

```python
outlier_z = (df['Population'] - df['Population'].mean()) /df['Population'].std()
dataz = df.loc[(outlier_z > 4) | (outlier_z < -4)]
print(dataz)
```

```python
outlier_z2 = (df['Income'] - df['Income'].mean()) /df['Income'].std()
dataz2 = df.loc[(outlier_z2 > 4) | (outlier_z2 < -4)]
print(dataz2)
```
